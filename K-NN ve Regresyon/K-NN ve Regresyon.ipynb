## K-NN Algoritması

```python
#Verisetini Giriş(X) ve cıkış(y) olarak ayıralım
X = df2.drop(["Species"], axis =1)
y = df2["Species"]

#X = df.iloc[:,[0,1,2,3]].values
#y = df.iloc[:,4].values
```

```python
X.head()
```

```python
y.head()
```

```python
#Veristini test ve train olarak ayıralım 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 101, test_size = 0.33)
```

random_state=101 parametresi, veri setinin rastgele bölünmesinde kullanılan rasgele durumun belirlenmesini sağlar. Bu, her çalıştırmada aynı bölünmeyi elde etmemize yardımcı olur.
test_size=0.33 parametresi, veri setinin test setine ayrılacak oranını belirler. Bu durumda, veri setinin %33'ü test seti olarak ayrılırken, geri kalan %67'si eğitim seti olarak kullanılır.

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

### Model Değerlendirmesi

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
accuracy = model.score(X_test, y_test)
print('Accuracy = '+ str(accuracy))
```

```python
predictions = model.predict(X_test)
print("confusion matrix:\n ",confusion_matrix(y_test, predictions))
```

```python
print("classification report:\n ",classification_report(y_test, predictions))
```
