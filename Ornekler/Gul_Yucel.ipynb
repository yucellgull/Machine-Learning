# Soru-1

```python
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Veri setini yükleme

```python
wine = load_wine()
data = pd.DataFrame(data=wine.data, columns=wine.feature_names)
data['target'] = wine.target
```

## Veri analizi

```python
data.head()
```

## Veri Seti İstatistikleri

```python
data.describe()
```

## Null değer kontrolü

```python
data.isnull().sum()
```

## Veri dağılımı inceleme


```python
sns.set(style="whitegrid")
for feature in data.columns[:-1]:
    sns.histplot(data[feature], kde=True, bins=20, color='skyblue')
    plt.title(feature)
    plt.show()
```

## Korelasyon ilişkisi inceleme


```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".1f", linewidths=0.5)
plt.title("Korelasyon Matrisi")
plt.show()
```

# Soru-2

```python
import random
```

## Rastgele en az 5 elemanlı bir sözlük oluşturma

```python
sozluk = {f"anahtar{i}": random.randint(0, 100) for i in range(5)}
sozluk
```

## Rastgele bir anahtar seçme

```python
secilen_anahtar = random.choice(list(sozluk.keys()))
```

## Seçilen anahtarı kullanarak sözlükten eleman silme

```python
del sozluk[secilen_anahtar]
```

```python
sozluk
```

## Yeni bir eleman ekleyerek sözlüğü güncelleme

```python
yeni_anahtar = "yeni_anahtar"
yeni_deger = random.randint(0, 100)
sozluk[yeni_anahtar] = yeni_deger
sozluk
```

## Mevcut bir elemanı güncelleme

```python
guncellenecek_anahtar = random.choice(list(sozluk.keys()))
yeni_deger = random.randint(0, 100)
sozluk[guncellenecek_anahtar] = yeni_deger
sozluk
```
