## Spektral Kümeleme Kullanarak Kredi Kartı Veri Kümelemesi 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import SpectralClustering
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
```

```python
import warnings
warnings.filterwarnings("ignore", "is_categorical_dtype")
warnings.filterwarnings("ignore", "use_inf_as_na")
warnings.filterwarnings("ignore", category=FutureWarning)
```

### verileri temizleme

```python
X = pd.read_csv('CC_GENERAL.csv')
 
# sütün silme
X = X.drop('CUST_ID', axis = 1)
 
# boş verileri doldurma
X.fillna(method ='ffill', inplace = True)
 
X.head()
```

### Verileri görselleştirilebilir hale getirmek için verilerin ön işlenmesi 

```python

 
# Verilerin Ölçeklenmesi
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
 
# Verilerin Normalizasyonu
X_normalized = normalize(X_scaled)
 
# NumPy Dizisinin Pandas DataFrame'e Dönüştürülmesi
X_normalized = pd.DataFrame(X_normalized)
 
# Boyut Azaltılmış Verilerin DataFrame Olarak Düzenlenmesi
pca = PCA(n_components = 2)
X_principal = pca.fit_transform(X_normalized)
X_principal = pd.DataFrame(X_principal)
X_principal.columns = ['P1', 'P2']
 
X_principal.head()
```

### Kümeleme modellerini oluşturma ve Kümelemeyi Görselleştirme 

```python
# Spektral Kümeleme Modelinin Oluşturulması
spectral_model_rbf = SpectralClustering(n_clusters = 2, affinity ='rbf')
 
# Modelin Eğitimi ve Küme Etiketlerinin Tahmini
labels_rbf = spectral_model_rbf.fit_predict(X_principal)
```

##### Aşağıdaki kod, spektral kümeleme ile elde edilen kümeleri görselleştirmek için bir dağılım grafiği oluşturmaktadır. Grafiğin renklendirilmesiyle, farklı kümelere ait veri noktaları ve bunların dağılımları görsel olarak anlaşılabilmektedir.

```python
# Etiket-Renk Eşlemesi Oluşturma
colours = {}
colours[0] = 'b'
colours[1] = 'r'
 
# Veri Noktaları için Renk Vektörü Oluşturma
cvec = [colours[label] for label in labels_rbf]
 
# Plotting the clustered scatter plot
 
b = plt.scatter(X_principal['P1'], X_principal['P2'], color ='b');
y = plt.scatter(X_principal['P1'], X_principal['P2'], color ='r');
 
plt.figure(figsize =(9, 9))
plt.scatter(X_principal['P1'], X_principal['P2'], c = cvec)
plt.legend((b, y), ('Label 0', 'Label 1'))
plt.show()
```

###  yakınlık = 'en yakın_komşular' 

##### Bu kod, spektral kümeleme yöntemini kullanarak kredi kartı verilerini tekrar iki kümeye ayırır. Ancak bu sefer, veri noktaları arasındaki benzerlikleri hesaplamak için farklı bir yöntem ("en yakın komşular") kullanılmaktadır. Bu, farklı şekil ve yoğunluklardaki kümeleri daha iyi ayırt etmeye yardımcı olabilir.

```python
# Spektral Kümeleme Modelinin Oluşturulması
spectral_model_nn = SpectralClustering(n_clusters = 2, affinity ='nearest_neighbors')
 
#  Modelin Eğitimi ve Küme Etiketlerinin Tahmini
labels_nn = spectral_model_nn.fit_predict(X_principal)
```

### Performansların değerlendirilmesi 

```python
#  Benzerlik Ölçütlerinin Listelenmesi
affinity = ['rbf', 'nearest-neighbours']
 

s_scores = []
 

s_scores.append(silhouette_score(X, labels_rbf))
s_scores.append(silhouette_score(X, labels_nn))
 
print(s_scores)
```

### Performansların karşılaştırılması 

```python
plt.bar(affinity, s_scores)
plt.xlabel('Affinity')
plt.ylabel('Silhouette Score')
plt.title('Comparison of different Clustering Models')
plt.show()
```
